```python
# Libraries
import requests
import json
import pandas as pd
```

```python
url_base = 'https://earthquake.usgs.gov/fdsnws/event/1/count?format=geojson'
contri = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2020-01-01&endtime=2020-12-31&minmagnitude=5'
```

```python
r = requests.get(contri)
```

```python
r.status_code
```

```python
r.json()
```

```python
r
```

```python
df = pd.DataFrame(r.json()['features'])
df.head()
```

```python
df1 = pd.json_normalize(df['properties'])
df1.head()
```

```python
df2 = pd.json_normalize(df['geometry'])
df2.head()
```

```python
# combinando dataframes de coordenadas y propiedades de los terremotos.
merge_df = pd.concat([df1, df2], axis=1)
```

```python
merge_df.head()
```

```python
tabla_salida = merge_df[['mag','place','time','updated','tz','alert','status','tsunami','rms','magType','type','coordinates']]
```

```python
tabla_salida.head()
```

```python
tabla_salida.columns = ['mag','place','time','updated','tz','alert','status','flag_tsunami','rms','magType','earthquake_type','coordinate_type','coordinates']
```
